```python
# for short video

##/media/flyranch/data21/magnotether/data/20230906/fly1/20230906133249.txt

##/media/flyranch/data21/magnotether/data/20230907/fly3/20230907150023.txt


#import json
import numpy as np
import os
import glob
import sys
import pandas as pd
import cv2
import matplotlib
from matplotlib import pyplot as plt

def get_LED_angle_in_rad(LED_num):
    ###2022/10/25
    ###2022/10/11
#                -180/+180 (LED 138)

#        -90(32)                    +90 (105)

#                    0 (68)

    LED_num=LED_num.astype("float")
    LED_num[(LED_num==-1)|(LED_num==149)|(LED_num==150)]=np.nan ### for dark mode
    LED_num[LED_num==68]=0
    LED_num[LED_num==105]=np.pi/2
    LED_num[LED_num==138]=np.pi
    LED_num[LED_num==32]=-np.pi/2
    LED_num[LED_num==14]=-3*np.pi/4
    LED_num[LED_num==50]=-np.pi/4
    LED_num[LED_num==87]=np.pi/4
    LED_num[LED_num==121]=3*np.pi/4    
    output_LED_arr=LED_num

    return output_LED_arr

def new_data_frame_maker(path):
### create new data frame by adding diff from text data
    df=pd.read_csv(path,delimiter=' ',header=None) ### for actual running experiment
#    df=pd.read_csv(path,delimiter='\t',header=None) ### for dummy
    df.columns=("time","heading","LED")

    ## remove rows when LED at -1, 149,150
    new_df=df[df.LED != -1]
#    new_df=new_df[new_df.LED != 150]
    new_df=new_df[new_df.LED != 149]

    ## for "heading", convert -90 to 270
#    new_df.loc[new_df['heading']==-90,'heading']=270

    LED_angles_in_deg=np.rad2deg(get_LED_angle_in_rad(new_df['LED']))
    diff=new_df['heading']-LED_angles_in_deg

    ## adjust diff
    low_inds=diff<-180
    diff[low_inds]=diff[low_inds]+360
    high_inds=diff>180
    diff[high_inds]=diff[high_inds]-360

    ## adding "diff" column, diff in degree
    new_df.insert(3,"diff",diff,True)

    return new_df

def vector_calc_from_degree(angle):
    rad=np.deg2rad(angle)
    V_x=np.cos(rad)
    V_y=np.sin(rad)
    return V_x, V_y

#def find_row_number(dataframe):
    ### return the row number at LED tuened on 
#    n=0
#    for i in range(len(dataframe)):
#    print(new_df.iloc[i,2])
#        if n!=dataframe.iloc[i,2]:
#            row_num=dataframe[dataframe['LED']==dataframe.iloc[i,2]].index
#            n=dataframe.iloc[i,2]
#    return len(row_num)    


#def mag_img_with_heading_maker(angle,img):
#    vx,vy=vector_calc_from_degree(angle)   
#    the_image=cv2.imread(img)
#    matplotlib.rcParams['pdf.fonttype'] = 42

#    fig,ax=plt.subplots(figsize=(10,10))
#    the_image=ax.imshow(the_image,extent=[0,1,0,1])

    ### origin decided by my eyes
#    ax.plot([0.43],[0.57],'or',markersize=10)
    ### 0 degree
    #ax.plot([0.43+0.4],[0.59],'<b',markersize=10)
    #ax.text(0.43+0.4,0.59-0.05,'LED69(0 degree)',color='b',fontsize=12)
    ### 180 degree
    #ax.plot([0.43-0.4],[0.59],'ob',markersize=10)
    #ax.text(0.43-0.4,0.59-0.05,'LED138 (180 degree)',color='b',fontsize=12)
    ### -90 degree
    #ax.plot([0.43],[0.59+0.4],'vb',markersize=10)
    #ax.text(0.43+0.05,0.59+0.4-0.02,'LED35(-90 degree)',color='b',fontsize=12)
    ### +90 degree
    #ax.plot([0.43],[0.59-0.4],'^b',markersize=10)
    #ax.text(0.43+0.05,0.59-0.4,'LED104(+90 degree)',color='b',fontsize=12)
    ### because right side is 0 degree, bottom is 90, left is 180, upper is -90, the center minus V_y needed
### devided by 3 for adjusting 
#    ax.plot([0.43,0.43+vx/3],[0.57,0.57-vy/3],'-or',linewidth=2,markersize=10)
#    plt.show()
#    return fig

def mag_img_with_heading_and_sun_maker(angle,img,LED):
    vx,vy=vector_calc_from_degree(angle)    
    the_image=cv2.imread(img)
    matplotlib.rcParams['pdf.fonttype'] = 42
    fig,ax=plt.subplots(figsize=(10,10))
    the_image=ax.imshow(the_image,extent=[0,1,0,1])

    ### origin decided by my eyes
    ax.plot([0.43],[0.57],'or',markersize=10)
    
    if LED==150:
        pass
    elif LED==14: #-135 degree
        ax.plot([0.43-0.4],[0.57+0.4],'oy',markersize=10)
        ax.text(0.43-0.4,0.57+0.4-0.05,'Sun',color='y',fontsize=12)
    elif LED==50: #-45 degree
        ax.plot([0.43+0.4],[0.57+0.4],'oy',markersize=10)
        ax.text(0.43+0.4,0.57+0.4-0.05,'Sun',color='y',fontsize=12)
    elif LED==87: #+45 degree
        ax.plot([0.43+0.4],[0.57-0.4],'oy',markersize=10)
        ax.text(0.43+0.4,0.57-0.4-0.05,'Sun',color='y',fontsize=12)
    elif LED==121: #+135 degree
        ax.plot([0.43-0.4],[0.57-0.4],'oy',markersize=10)
        ax.text(0.43-0.4,0.57-0.4-0.05,'Sun',color='y',fontsize=12)
### devided by 3 for adjusting 
    ax.plot([0.43,0.43+vx/3],[0.57,0.57-vy/3],'-or',linewidth=2,markersize=10)
    return fig

data_path='/media/flyranch/data21/magnotether/mag_20230907150023_ref/'
text_path=data_path+'20230907150023.txt'
df=new_data_frame_maker(text_path)
# import all image files with the .jpg extension
images=glob.glob(data_path+'20230907150023_images/*.png')
new_df=df.loc[df['time']<=df['time'].iloc[0]+240] #240secs >> 1st dark, 1st LED, 2nd LED periods
sorted_images=sorted(images)[6:6+len(new_df)] # since the first 6 rows data on LED -1 

out_path=data_path+'annotated_20230907150023_images/'

count=1
ext='.png'
#row_num=find_row_number(new_df)


for i in sorted_images:
    img=i
    angle=new_df.iloc[sorted_images.index(i),1]
    led=new_df.iloc[sorted_images.index(i),2]
    mag_img_with_heading_and_sun_maker(angle,img,led)
    if len(str(count))==1:
        plt.savefig(out_path+'annotated_20230907150023_000'+str(count)+ext,transparent=True)
    elif len(str(count))==2:
        plt.savefig(out_path+'annotated_20230907150023_00'+str(count)+ext,transparent=True)
    elif len(str(count))==3:
        plt.savefig(out_path+'annotated_20230907150023_0'+str(count)+ext,transparent=True)
    elif len(str(count))==4:
        plt.savefig(out_path+'annotated_20230907150023_'+str(count)+ext,transparent=True)
    count+=1



#tst_angle=new_df.iloc[0,1]
#tst_img=sorted_images[0]
#print(tst_angle)
#print(tst_img)

#tst_fig=mag_img_with_heading_maker(tst_angle,tst_img)

#print(type(tst_fig))

#out_path='/media/flyranch/data2/mag_20220803_ref/'
#cv2.imwrite(out_path+'tst.png',tst_fig)
#plt.savefig(out_path+'tst.pdf',transparent=True)
###note 0006==row5


###average image

#image_data=[]
#for img in sorted_images:
#    this_image=cv2.imread(img,1)
#    image_data.append(this_image)

#avg_image=image_data[0]
#for i in range(len(image_data)):
#    if i==0:
#        pass
#    else:
#        alpha=1.0/(i + 1)
#        beta=1.0 - alpha
#        avg_image=cv2.addWeighted(image_data[i],alpha,avg_image,beta,0.0)
       
#out_path='/media/flyranch/data2/mag_20220803_ref/'      
#cv2.imwrite(out_path+'avg.png',avg_image)
#avg_image=cv2.imread(out_path+'avg.png')

#the_image=cv2.imread(sorted_images[0])
#fig,ax=plt.subplots(figsize=(10,10))
#the_image=ax.imshow(the_image,extent=[0,1,0,1])

### origin decided by my eyes
#ax.plot([0.43],[0.59],'or',markersize=10)

#ax.plot([0.43+0.4],[0.59],'<b',markersize=10)
#ax.text(0.43+0.4,0.59-0.05,'LED69(0 degree)',color='b',fontsize=12)

#ax.plot([0.43-0.4],[0.59],'>b',markersize=10)
#ax.text(0.43-0.4,0.59-0.05,'LED138(180 degree)',color='b',fontsize=12)

#ax.plot([0.43],[0.59+0.4],'vb',markersize=10)
#ax.text(0.43+0.05,0.59+0.4-0.02,'LED35(-90 degree)',color='b',fontsize=12)

#ax.plot([0.43],[0.59-0.4],'^b',markersize=10)
#ax.text(0.43+0.05,0.59-0.4,'LED104(+90 degree)',color='b',fontsize=12)

#deg=new_df.iloc[0,1]

#V_x,V_y=vector_calc_from_degree(deg)

#print(V_x)
#print(V_y)


### because right side is 0 degree, bottom is 90, left is 180, upper is -90, the center minus V_y needed
### devided by 3 for adjusting 
#ax.plot([0.43,0.43+V_x/3],[0.59,0.59-V_y/3],'-or',linewidth=2,markersize=10)
#plt.show()
```

```python
# for short video
### esp, annotated images

##/media/flyranch/data21/magnotether/data/20230920/fly5/20230920152326.txt

import json
import numpy as np
import os
import glob
import sys
import pandas as pd
import cv2
import matplotlib
from matplotlib import pyplot as plt
plt.rcParams.update({'figure.max_open_warning': 0}) ### to avoid "max open warning"
matplotlib.rcParams['pdf.fonttype'] = 42

#text=input('which date data?: ')
text='20230920152326'
#Y_N=input('Is this the first run?(y or n): ')
Y_N='y'

def get_LED_angle_in_rad(LED_num):
    ###2022/10/25
    ###2022/10/11
    ###2023/09/26
#                -180/+180 (LED 138)
#             -135 (121)      +135 (14)    
#        -90(105)                    +90 (32)
#              -45 (87)      +45 (50)
#                    0 (68)

    LED_num=LED_num.astype("float")
    LED_num[(LED_num==-1)|(LED_num==149)|(LED_num==150)]=np.nan ### for dark mode
    LED_num[LED_num==68]=0
    LED_num[LED_num==105]=-np.pi/2
    LED_num[LED_num==138]=np.pi
    LED_num[LED_num==32]=np.pi/2
    ## since 20230920
    LED_num[LED_num==14]=3*np.pi/4 
    LED_num[LED_num==50]=np.pi/4
    LED_num[LED_num==87]=-np.pi/4
    LED_num[LED_num==121]=-3*np.pi/4
    output_LED_arr=LED_num

    return output_LED_arr

def new_data_frame_maker(path):
### create new data frame by adding diff from text data
    df=pd.read_csv(path,delimiter=' ',header=None) ### for actual running experiment
#    df=pd.read_csv(path,delimiter='\t',header=None) ### for dummy
    df.columns=("time","heading","LED")

    ## remove rows when LED at -1, 149
    new_df=df[df.LED != -1]
    new_df=new_df[new_df.LED != 149]

    LED_angles_in_deg=np.rad2deg(get_LED_angle_in_rad(new_df['LED']))
    diff=new_df['heading']-LED_angles_in_deg

    ## adjust diff
    low_inds=diff<-180
    diff[low_inds]=diff[low_inds]+360
    high_inds=diff>180
    diff[high_inds]=diff[high_inds]-360

    ## adding "diff" column, diff in degree
    new_df.insert(3,"diff",diff,True)

    return new_df

def vector_calc_from_degree(angle):
    rad=np.deg2rad(angle)
    V_x=np.cos(rad)
    V_y=np.sin(rad)
    return V_x, V_y

def mag_img_with_heading_and_sun_maker(angle,img,LED):
    vx,vy=vector_calc_from_degree(angle)    
    the_image=cv2.imread(img)
    fig,ax=plt.subplots(figsize=(6,6))
    plt.axis('off')
    the_image=ax.imshow(the_image,extent=[0,1.2,0,1.2])
    #plt.ioff()
    #the_image=ax.plot(the_image,extent=[0,1.2,0,1.2])
    ### origin decided by my eyes
    ax.plot([0.83],[0.5],'or',markersize=10)
    #ax.spines[['bottom','right','top','left']].set_visible(False)
    
    if LED==150:
        pass
    elif LED==14: #+135 degree
        ax.plot([0.83-0.35],[0.5-0.35],'oy',markersize=10)
        ax.text(0.83-0.35,0.5-0.35-0.05,'Sun',color='y',fontsize=12)
    elif LED==50: #+45 degree
        ax.plot([0.83+0.35],[0.5-0.35],'oy',markersize=10)
        ax.text(0.83+0.35,0.5-0.35-0.05,'Sun',color='y',fontsize=12)
    elif LED==87: #-45 degree
        ax.plot([0.83+0.35],[0.5+0.35],'oy',markersize=10)
        ax.text(0.83+0.35,0.5+0.35-0.05,'Sun',color='y',fontsize=12)
    elif LED==121: #-135 degree
        ax.plot([0.83-0.35],[0.5+0.35],'oy',markersize=10)
        ax.text(0.83-0.35,0.5+0.35-0.05,'Sun',color='y',fontsize=12)
### devided by 3 for adjusting 
    ax.plot([0.83,0.83+vx/3],[0.5,0.5-vy/3],'-or',linewidth=2,markersize=10)
    return fig

data_path='/media/flyranch/data21/magnotether/mag_'+text+'_ref/'
text_path=data_path+text+'.txt'
df=new_data_frame_maker(text_path)
index_vals=df[df['LED']==150]
val=index_vals.index.tolist()[0] # this value show number of -1 LED 
# import all image files with the .png extension
images=glob.glob(data_path+text+'_images/*.png')
if Y_N=='y':
    new_df=df.loc[df['time']<=df['time'].iloc[0]+240] #240secs >> 1st dark, 1st LED, 2nd LED periods
                                                    # if just set 240, PC crashes because of huge memory usage
                                                    #it cannot be done just once
elif Y_N=='n':
    new_df=df.loc[(df['time'].iloc[0]+210<df['time'])&(df['time']<=df['time'].iloc[0]+240)]
                                                    # 30 sec chunks recommended
    
#print(new_df)
initial_row_num=new_df.index.tolist()[0] 
#print(initial_row_num)
if Y_N=='y':
    initial_row_num=val
    sorted_images=sorted(images)[val:val+len(new_df)]
elif Y_N=='n':
    sorted_images=sorted(images)[initial_row_num:initial_row_num+len(new_df)]
out_path=data_path+'annotated_'+text+'_images/'

count=1+initial_row_num-val
ext='.png'
#ext='.svg'

for i in sorted_images:
    img=i
    angle=new_df.iloc[sorted_images.index(i),1]
    led=new_df.iloc[sorted_images.index(i),2]
    mag_img_with_heading_and_sun_maker(angle,img,led)
    if len(str(count))==1:
        #plt.savefig(out_path+'annotated_'+text+'_000'+str(count)+ext,transparent=True)
        plt.savefig(out_path+'annotated_'+text+'_000'+str(count)+ext,dpi=300)
    elif len(str(count))==2:
        #plt.savefig(out_path+'annotated_'+text+'_00'+str(count)+ext,transparent=True)
        plt.savefig(out_path+'annotated_'+text+'_00'+str(count)+ext,dpi=300)
    elif len(str(count))==3:
        #plt.savefig(out_path+'annotated_'+text+'_0'+str(count)+ext,transparent=True)
        plt.savefig(out_path+'annotated_'+text+'_0'+str(count)+ext,dpi=300)
    elif len(str(count))==4:
        #plt.savefig(out_path+'annotated_'+text+'_'+str(count)+ext,transparent=True)
        plt.savefig(out_path+'annotated_'+text+'_'+str(count)+ext,dpi=300)
    #print('row '+ str(val+count)+' heading angle: '+str(angle)) #to check it with original text data
    count+=1
    plt.close('all') #close plot each time
```

```python
from __future__ import print_function
import numpy as np
import cv2
import pdb


def find_fly_angle(image, threshold=60, mask_scale=0.95): 
    """
    Finds the angle of the fly given an image.  Fly should be dark against a light background.

    Parameters:

      threshold   = value of threshold used to separate fly's from the (light) image background.

      mask_scale  = scales the radius of the circular mask around the fly's centroid. The value 
                    Should be in range [0,1]. 1 means the radius = min(image_width, image_height) 

    Returns: tuple (angle, angle_data) where

      angle      = the estimate of the fly's angle in radians

      angle_data = dictionary of useful data and images calculated during the angle estimation

      angle_data = { 
          'flipped': boolean which indicates whether or not fly's orientation was flipped,
          'moments': the moments of the maximum area contour in the thresholded image,
          'max_contour': the maximum area contour,
          'max_contour_area':  the area of the maximum area contour,
          'body_vector': unit vector along the fly's body axis,
          'contour_image': image with body contour, centroid, and fly's body axis drawn on it,
          'shifted_image': image shifted so that the fly's cendroid is at the center,
          'rotated_image': image rotated by the fly's body angle an shift so centroid is centered,
          'threshold_image': the thresholded image,
          'shifted_threshold_image': thresholded image shifted so the centroid is centered,
          'rotated_threshold_image': thresholded image rotated and shifted 
      }

    """

    # Get basic image data
    
    height, width = image.shape
    image_cvsize = width, height 
    mid_x, mid_y = 0.5*width, 0.5*height

    # Threshold, find contours and get contour with the maximum area
    rval, threshold_image = cv2.threshold(image, threshold, np.iinfo(image.dtype).max, cv2.THRESH_BINARY_INV)
    
    dummy, contour_list, dummy = cv2.findContours(threshold_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    max_contour, max_area = get_max_area_contour(contour_list)

    # Get moments and then compute centroid and rotation angle
    moments = cv2.moments(max_contour)
    centroid_x, centroid_y = get_centroid(moments)
    angle, body_vector = get_angle_and_body_vector(moments)

    # Get bounding box and find diagonal - used for drawing body axis
    bbox = cv2.boundingRect(max_contour)
    bbox_diag = np.sqrt(bbox[2]**2 + bbox[3]**2)

    # Create points for drawing axis fly in contours image 
    axis_length = 0.75*bbox_diag
    body_axis_pt_0 = int(centroid_x + axis_length*body_vector[0]), int(centroid_y + axis_length*body_vector[1])
    body_axis_pt_1 = int(centroid_x - axis_length*body_vector[0]), int(centroid_y - axis_length*body_vector[1])

    # Compute cirlce mask
    mask_radius = int(mask_scale*height/2.0)
    vals_x = np.arange(0.0,width)
    vals_y = np.arange(0.0,height)
    grid_x, grid_y = np.meshgrid(vals_x, vals_y)
    circ_mask = (grid_x - width/2.0 + 0.5)**2 + (grid_y - height/2.0 + 0.5)**2 < (mask_radius)**2

    # Draw image with body contours, centroid circle and body axis
    centroid = int(centroid_x), int(centroid_y)
    contour_image = cv2.cvtColor(image,cv2.COLOR_GRAY2BGR)
    cv2.drawContours(contour_image,[max_contour],-1,(0,0,255),2)
    cv2.circle(contour_image, centroid, 5, (0,0,255), 2)
    cv2.line(contour_image, body_axis_pt_0, body_axis_pt_1, (0,0,255), 2)
    cv2.circle(contour_image, centroid, mask_radius, (0,0,255),2) 

    # Get matrices for shifting (centering) and rotating the image
    shift_mat = np.matrix([[1.0, 0.0, (mid_x - centroid_x)], [0.0, 1.0, (mid_y - centroid_y)]]) 
    rot_mat = cv2.getRotationMatrix2D((mid_x, mid_y),np.rad2deg(angle),1.0)

    # Shift and rotate the original image
    shifted_image = cv2.warpAffine(image, shift_mat, image_cvsize)
    rotated_image = cv2.warpAffine(shifted_image,rot_mat,image_cvsize)

    # Shift and rotate threshold image. 
    shifted_threshold_image = cv2.warpAffine(threshold_image, shift_mat, image_cvsize)
    rotated_threshold_image = cv2.warpAffine(shifted_threshold_image,rot_mat,image_cvsize)
    rotated_threshold_image = rotated_threshold_image*circ_mask
    rval, rotated_threshold_image = cv2.threshold(rotated_threshold_image, threshold, np.iinfo(image.dtype).max, cv2.THRESH_BINARY)
    rotated_threshold_image_pre=rotated_threshold_image
    # Get orientation discriminant and flip image if needed 
    orient_ok, orient_discrim, orient_dict = is_orientation_ok(rotated_threshold_image,2)
    
    if not orient_ok:
        rot_180_mat = cv2.getRotationMatrix2D((mid_x, mid_y),-180.0,1.0)
        rotated_image = cv2.warpAffine(rotated_image,rot_180_mat,image_cvsize)
        rotated_threhold_image = cv2.warpAffine(rotated_threshold_image,rot_180_mat,image_cvsize)
        angle += np.deg2rad(-180.0)

    angle = normalize_angle_range(angle)

    data = {
            'flipped': not orient_ok,
            'moments': moments,
            'max_contour': max_contour,
            'max_contour_area':  max_area,
            'body_vector': body_vector,
            'contour_image': contour_image,
            'shifted_image': shifted_image,
            'rotated_image': rotated_image,
            'threshold_image': threshold_image,
            'shifted_threshold_image': shifted_threshold_image,
            'rotated_threshold_image': rotated_threshold_image,
            }

    return angle, data, orient_dict
        

def is_orientation_ok(image,k=2,is_first=True): 
    """
    Returns True if orienation is OK and False if fly need to be flipped. 

    Takes thresholded, shifted and rotated image and computes an orientation
    discriminant which is is the ratio of the k-th moments of area for each
    half of the body.  The discriminant should be such that it is > 1 for one
    orientation and < 1 for the other. 

    """
    orientation_dict={}
    mid_x, mid_y = int(0.5*image.shape[1]), int(0.5*image.shape[0])

    # Get moment for first body half 
    image_0 = np.array(image)
    image_0[:,:int(mid_x)] = 0
    image_0 = image_0[:,int(mid_x):]
    moment_0 = get_moment(image_0,k)

    # Get moment for second body half
    image_1 = np.array(image)
    image_1[:,int(mid_x):] = 0
    image_1copy=np.copy(image_1)
    image_1 = np.fliplr(image_1)
    image_1 = image_1[:,int(mid_x):]
    moment_1 = get_moment(image_1,k)
    # Compute descriminant and flip flag
    discrim = (moment_0 - moment_1)/(moment_0 + moment_1)
    if discrim < 0:
        ok = False
    else:
        ok = True 
    

    dummy, contour_list0, dummy = cv2.findContours(image_0, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    max_contour, max_area = get_max_area_contour(contour_list0)
    perimeter_0 = cv2.arcLength(max_contour,True)

    dummy, contour_list1, dummy = cv2.findContours(image_1copy, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    max_contour, max_area = get_max_area_contour(contour_list1)
    perimeter_1 = cv2.arcLength(max_contour,True)

    perimeter_discrim=(perimeter_0-perimeter_1)/(perimeter_0+perimeter_1)
    orientation_dict['image_0']=image_0
    orientation_dict['image_1']=image_1
    orientation_dict['moment_0']=moment_0
    orientation_dict['moment_1']=moment_1
    orientation_dict['perimeter_0']=perimeter_0
    orientation_dict['perimeter_1']=perimeter_1
    orientation_dict['discrim']=discrim
    orientation_dict['perimeter_discrim']=perimeter_discrim
    return ok, discrim, orientation_dict

def calculate_moments(image):
    # return six images and 12 moments (first, second)
    pass

def get_moment(image,k=2):
    
    data = image.sum(axis=0)
    weighted_data = data*np.arange(data.shape[0])**k
    return  weighted_data.sum()


def get_max_area_contour(contour_list):
    """
    Given a list of contours finds the contour with the maximum area and 
    returns 
    """
    contour_areas = np.array([cv2.contourArea(c) for c in contour_list])
    max_area = contour_areas.max()
    max_ind = contour_areas.argmax()
    max_contour = contour_list[max_ind]
    return max_contour, max_area


def get_centroid(moments): 
    """
    Computer centroid given the image/blob moments
    """
    if moments['m00'] > 0:
        centroid_x = moments['m10']/moments['m00']
        centroid_y = moments['m01']/moments['m00']
    else:
        centroid_x = 0.0
        centroid_y = 0.0
    return centroid_x, centroid_y


def get_angle_and_body_vector(moments): 
    """
    Computre the angle and body vector given the image/blob moments
    """
    body_cov = np.array( [ [moments['mu20'], moments['mu11']], [moments['mu11'], moments['mu02'] ]])
    eig_vals, eig_vecs = np.linalg.eigh(body_cov)
    max_eig_ind = np.argmax(eig_vals**2)
    max_eig_vec = eig_vecs[:,max_eig_ind]
    angle = np.arctan2(max_eig_vec[1], max_eig_vec[0])
    return angle, max_eig_vec


def normalize_angle_range(angle):
    """
    Normalize the angle value so that is between -pi and pi
    """
    angle_adj = angle
    angle_adj += np.deg2rad(180.0)
    angle_adj = angle_adj % np.deg2rad(360.)
    angle_adj -= np.deg2rad(180.0)
    return angle_adj
```
